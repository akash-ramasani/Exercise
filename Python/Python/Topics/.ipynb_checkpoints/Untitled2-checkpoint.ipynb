```python
import os
import re

def rename_files(directory):
    # Walk through all subdirectories and files in the directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.dat'):
                # Extracting components from the filename
                match = re.match(r'PJTSAU_(\d+)_(\d{4})_(\d{2})_(\d{2})_(\d{4})\.dat', file)
                if match:
                    number = match.group(1)
                    year = match.group(2)
                    month = match.group(3)
                    day = match.group(4)
                    
                    # Renaming the file
                    new_filename = f'PJTSAU_{year}-{month}-{day}_{match.group(5)}.dat'
                    new_path = os.path.join(root, new_filename)
                    
                    # Check if the new filename already exists
                    if os.path.exists(new_path):
                        # If the new filename already exists, add a counter to make it unique
                        counter = 1
                        while os.path.exists(new_path):
                            new_filename = f'PJTSAU_{year}-{month}-{day}_{match.group(5)}_{counter}.dat'
                            new_path = os.path.join(root, new_filename)
                            counter += 1
                    
                    old_path = os.path.join(root, file)
                    os.rename(old_path, new_path)
                    print(f'Renamed {old_path} to {new_path}')

# Change 'directory_path' to the path of the directory containing the folders
directory_path = r"C:\Users\PC-14\Desktop\Kharif 2023"
rename_files(directory_path)
```
