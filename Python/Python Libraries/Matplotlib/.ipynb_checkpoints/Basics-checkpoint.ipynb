```python
import matplotlib.pyplot as plt

#We can also import matplot.pylot as 'from matplotlib import pyplot as plt'
```

```python
import numpy as np
```

```python
x_points = np.array([2, 5, 9, 12, 1])
y_points = np.array([17, 15, 2, 16, 7])

plt.plot(x_points, y_points) # plot() function is used to draw points in a diagram
plt.show()
```

Plotting Without Line

```python
plt.plot(x_points, y_points, 'o')
plt.show()
```

Default X-Points

```python
plt.plot(y_points) #If we did not specify the x-points, by default we will get 0, 1, 2, 3 default values.
plt.show()
```
